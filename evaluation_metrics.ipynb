# 📊 Understanding Evaluation Metrics in Machine Learning

This notebook explains key evaluation metrics, providing both textual definitions and visual examples of good and bad graphs.

## ✅ Confusion Matrix (Normalized) 
**Definition:** A normalized confusion matrix shows the proportion of predictions for each class instead of raw counts, making it easier to interpret model performance.

## ✅ Precision Curve
**Definition:** The precision curve plots the precision (positive predictive value) at different classification thresholds. Higher precision means fewer false positives.

### Good Example ✅
![](precision_curve_good.png)

### Bad Example ❌
![](precision_curve_bad.png)


## ✅ Recall Curve
**Definition:** The recall curve shows how well the model captures positive instances at different thresholds. Higher recall means fewer false negatives.

### Good Example ✅
![](recall_curve_good.png)

### Bad Example ❌
![](recall_curve_bad.png)


## ✅ F1 Score Curve
**Definition:** The F1 score curve balances precision and recall by computing their harmonic mean. A higher F1 score indicates a better balance between precision and recall.
